## Data Cleaning & Feature Engineering


```python
# convert the 'Date' column to datetime format
df['Date'] = pd.to_datetime(df['Date'], errors='coerce')

# extract 'Year', 'Month', 'Month_Name', 'Year_Month'
df['Year'] = df['Date'].dt.year
df['Month'] = df['Date'].dt.month
df['Month_Name'] = df['Date'].dt.month_name()
df['Year_Month'] = df['Date'].dt.to_period('M').astype(str) 

# extract hour and minute as 'Time_of_the_day'
df['Time_of_the_day'] = df['Date'].dt.strftime('%H:%M')

# convert 'Time' column to timedelta
df['Time'] = pd.to_timedelta(df['Time'], errors='coerce')

# extract day of the week
df['Day_of_Week'] = df['Date'].dt.day_name()

# add a column to indicate if the run was on a weekend
df['is_weekend'] = df['Date'].dt.dayofweek >= 5  # Saturday=5, Sunday=6
```

```python
# columns to convert data type to numeric
columns_to_convert = [
    'Distance', 'Calories', 'Avg HR', 'Max HR', 'Aerobic TE', 'Total Ascent', 
    'Total Descent', 'Steps', 'Avg Resp', 'Min Resp', 'Max Resp', 
    'Min Elevation', 'Max Elevation'
]

# apply the conversion to all specified columns
for column in columns_to_convert:
    df[column] = pd.to_numeric(df[column], errors='coerce')

# replace 0 in the 'Distance' column with NaN
df['Distance'] = df['Distance'].replace(0, float('nan'))
```

```python
# calculate longest running streak based on consecutive calendar days

# sort by date
running_activities = running_activities.sort_values('Date').reset_index(drop=True)

# calculate the difference in days between consecutive runs
running_activities['date_diff'] = running_activities['Date'].dt.date.diff().dt.days

# identify breaks in streaks where the gap is more than 1 day
running_activities['Streak'] = (running_activities['date_diff'] > 1).cumsum()

# count the number of days in each streak
streak_counts = running_activities.groupby('Streak').size()

# find the length of the longest streak
longest_streak = streak_counts.max()

# get the streak group with the longest run
longest_streak_id = streak_counts.idxmax()
longest_streak_info = running_activities[running_activities['Streak'] == longest_streak_id]

# get start and end date of the longest streak
streak_start = longest_streak_info['Date'].min().date()
streak_end = longest_streak_info['Date'].max().date()

print(f"The longest running streak was {longest_streak} days.")
print(f"It was between {streak_start} and {streak_end}.")
```

```python
# calculate the longest break between runs
# get the longest break
longest_break = running_activities['date_diff'].max()

# find the index where the longest break occurred
break_index = running_activities['date_diff'].idxmax()

# get the dates surrounding the longest break
before_break = running_activities.loc[break_index - 1, 'Date']
after_break = running_activities.loc[break_index, 'Date']

print(f"The longest break from running was {int(longest_break)} days.")
print(f"It was between {before_break.date()} and {after_break.date()}.")
```

```python
# extract the hour from the existing DataFrame
running_activities['Hour'] = running_activities['Date'].dt.hour

# count how many runs happened at each hour
hour_counts = running_activities['Hour'].value_counts().sort_index()

# ensure all hours (0-23) are represented, fill missing with 0
hour_counts = hour_counts.reindex(range(24), fill_value=0)

# find the index of the highest bar
highest_bar = hour_counts.idxmax()

# find the index of the lowest non-zero value
# filter out the zeros and find the minimum non-zero value
non_zero_hour_counts = hour_counts[hour_counts > 0]
lowest_bar = non_zero_hour_counts.idxmin()

# colors for higher, lowest and the rest of the values
colors = ['#63bb80' if hour == highest_bar else '#f98797' if hour == lowest_bar else '#87ceec' for hour in hour_counts.index]

# plot
plt.figure(figsize=(10, 6))

# plot the bar chart with the color customization
plt.bar(hour_counts.index, hour_counts.values, color=colors, edgecolor='black')

# title and labels
plt.title('Distribution of Runs by Hour of the Day')
plt.xlabel('Hour of the Day')
plt.ylabel('Number of Runs')

# customize the x-ticks to show all hours (0 to 23)
plt.xticks(range(0, 24))

# Add a grid
plt.grid(axis='y', linestyle='--', alpha=0.7)

# tighten the layout
plt.tight_layout()

# show the plot
plt.show()

# most frequent hour
most_common_hour = hour_counts.idxmax()
print(f"Most common time is between {most_common_hour}:00 and {most_common_hour}:59")

# least frequent hour
least_common_hour = non_zero_hour_counts.idxmin()
print(f"Least common time is between {least_common_hour}:00 and {least_common_hour}:59")
```

```python
# define the correct order of weekdays
ordered_days = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']

# count runs by day of the week and reorder
day_counts = running_activities['Day_of_Week'].value_counts().reindex(ordered_days)

# determine most and least common days
most_common_day = day_counts.idxmax()
non_zero_day_counts = day_counts[day_counts > 0]
least_common_day = non_zero_day_counts.idxmin()

# set colors
colors = ['#63bb80' if day == most_common_day else '#f98797' if day == least_common_day else '#87ceec' for day in day_counts.index]

# plot
plt.figure(figsize=(10, 6))
plt.bar(day_counts.index, day_counts.values, color=colors, edgecolor='black')

# styling
plt.title('Distribution of Runs by Day of the Week')
plt.xlabel('Day of the Week')
plt.ylabel('Number of Runs')
plt.xticks(rotation=0)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.ylim(0, day_counts.max() + 20)

# add value labels on bars
for i, value in enumerate(day_counts):
    plt.text(i, value + 1, str(int(value)), ha='center', va='bottom', fontsize=10)

plt.tight_layout()
plt.show()

# print results
print(f"Most common day: {most_common_day}")
print(f"Least common day: {least_common_day}")
```

## Clustering

```python
# select relevant features
features = ['Distance', 'Avg Speed', 'Avg HR', 'Total Ascent', 'Calories', 'Time']
clustering_data = running_activities[features].copy()
```

```python
clustering_data.dtypes
```

```python
# convert pace string (mm:ss) to total minutes as float
def pace_to_minutes(pace_str):
    try:
        minutes, seconds = map(int, str(pace_str).split(':'))
        return minutes + seconds / 60
    except:
        return None

clustering_data['Avg Speed'] = clustering_data['Avg Speed'].apply(pace_to_minutes)

# convert Time from timedelta to total minutes
clustering_data['Time'] = clustering_data['Time'].dt.total_seconds() / 60
```

```python
# check for missing values in the clustering data
missing_data = clustering_data.isnull().sum()

# display missing values for each column
print(missing_data)
```

For the purpose of this analysis, we will drop the rows where 'Avg HR' and 'Total Ascent' are missing.
Normally, we would explore other strategies to handle these missing values (e.g., imputation, using the median, etc.). However, the majority of the activities with missing values belong to 'Treadmill runs'. Therefore, removing these entries is a reasonable choice to maintain data consistency and analysis quality.

```python
# drop missing values
clustering_data = clustering_data.dropna(subset=['Avg HR', 'Total Ascent'])
```

```python
# initialize the scaler
scaler = StandardScaler()

# apply scaling to our clustering data
scaled_data = scaler.fit_transform(clustering_data)
```

```python
# try different numbers of clusters and record the inertia
inertia = []
K_range = range(1, 11)
for k in K_range:
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
    kmeans.fit(scaled_data)
    inertia.append(kmeans.inertia_)

# plot the elbow curve
plt.figure(figsize=(10, 6))
plt.plot(K_range, inertia, marker='o')
plt.title('Elbow Method for Optimal k')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Inertia')
plt.xticks(K_range)
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# apply KMeans with k=3
kmeans = KMeans(n_clusters=3, random_state=42, n_init=10)
clustering_data['Cluster'] = kmeans.fit_predict(scaled_data)
```

```python
# get average values per cluster
cluster_summary = clustering_data.groupby('Cluster').mean(numeric_only=True)

# round for readability
cluster_summary = cluster_summary.round(2)

# display the summary
print(cluster_summary)
```

```python
# add cluster labels to dataset for plotting
clustered_data = clustering_data.copy()
clustered_data['Cluster'] = kmeans.labels_

# define cluster titles
cluster_titles = {
    0: "Easy Runs - Low Intensity",
    1: "Moderate Runs - Medium Intensity",
    2: "Hard Runs - High Intensity"
}

# plot Distance vs Avg Speed
plt.figure(figsize=(10, 6))
for cluster in clustered_data['Cluster'].unique():
    subset = clustered_data[clustered_data['Cluster'] == cluster]
    plt.scatter(subset['Distance'], subset['Avg Speed'], 
                label=f'{cluster_titles[cluster]}', alpha=0.7)

plt.title('Clusters of Running Activities', fontsize=14, pad=20)
plt.xlabel('Distance (km)', fontsize=12)
plt.ylabel('Average Speed (min/km)', fontsize=12)
plt.legend(title='Run Types', fontsize=10)
plt.grid(True, linestyle='--', alpha=0.3)

# add some padding to prevent label cutoff
plt.tight_layout()

# show the plot
plt.show()
```
